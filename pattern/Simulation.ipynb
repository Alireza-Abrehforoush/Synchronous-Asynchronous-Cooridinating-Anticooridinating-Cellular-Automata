```python
import numpy as np
import regex as re
import random as rn
```

```python
def countOccurrences(string, substring):
    cnt = 0
    x = string + string
    m = len(substring)
    for i in range(len(string)):
        if substring == x[i: i + m]:
            cnt += 1
    return cnt


def countPatterns(string, patterns):
    n = len(patterns)
    result = np.zeros(n)
    for i in range(n):
        result[i] = countOccurrences(string, patterns[i])
    return result

def updateRule(string, i):
    n = len(string)
    if (i < n):
        if (string[i] != string[(i - 1) % n] and string[i] != string[(i + 1) % n]):
            pass
        else:
            list1 = list(string)
            list1[i] = str(1 - int(string[i]))
            string = "".join(list1)
    else:
        pass
    return string

def timeStep(string, threshold, patterns):
    #string_old = string
    r = np.random.random(size = len(string))
    for i in range(len(string)):
        if (r[i] <= threshold):
            string = updateRule(string, i)
    diff = countPatterns(string, patterns)# - countPatterns(string_old, patterns)
    return string, diff


def randKey(p):
    key = ""
    for i in range(p):
        temp = str(np.random.choice(["0", "1"]))
        key += temp
    return key

def getPotential(diff, alpha):
    return np.dot(diff, alpha.T)
```

```python
def main():
    str1 = randKey(1000)
    find = ["00", "11", "000", "001", "011", "100", "110", "111", "0000", "0001", "0010", "0011", "0100", "0110", "0111", "1000", "1001", "1011", "1100", "1101", "1110", "1111"]
    #alpha = np.ones(len(find)) / len(find)
    weights = np.array([2, 2, 4, 2, 2, 2, 2, 4, 6, 4, 2, 4, 2, 2, 4, 4, 2, 2, 4, 2, 4, 6])
    alpha = weights / np.sum(weights)
    print(str1)
    for i in range(10000):
        str2, diff = timeStep(str1, 0.95, find)
        print(getPotential(diff, alpha))
        str1 = str2
    print(str1)
    print(countPatterns(str1, find))
main()
```
